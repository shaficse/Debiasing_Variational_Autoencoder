```python
#Download Training Data
data_link = 'https://www.dropbox.com/s/hlz8atheyozp1yx/train_face.h5?dl=1'
path_to_training_data = tf.keras.utils.get_file('train_face.h5',data_link)
#Loading data 
loader = data_loader.TrainingDatasetLoader(path_to_training_data) 
```

```python
#We can look at the size of the training dataset and grab a batch of size 100:
number_of_training_example = loader.get_train_size()
(images, labels) = loader.get_batch(100)
```

```python
### Examining the CElebA training dataset
face_images = images[np.where(labels==1)[0]]
```

```python
not_face_images = images[np.where(labels==1)[0]]
```

```python
#@title Change the sliders to look at positive and negative training examples! { run: "auto" }
import random
idx_face = random.randint(0,len(face_images)) #@param {type:"slider", min:0, max:50, step:1}
idx_not_face = random.randint(0,len(not_face_images)) #@param {type:"slider", min:0, max:50, step:1}
print(idx_face, idx_not_face)
plt.figure(figsize=(5,5))
plt.subplot(1, 2, 1)
plt.imshow(face_images[idx_face])
plt.title("Face"); plt.grid(False)

plt.subplot(1, 2, 2)
plt.imshow(not_face_images[idx_not_face])
plt.title("Not Face"); plt.grid(False)
```
